```python
# import json
df["reactions"]

reactions_list = {
    "thÃ­ch" : [],
    "thÆ°Æ¡ng thÆ°Æ¡ng" : [],
    "yÃªu thÃ­ch" : [],
    "haha" : [],
    "buá»“n" : [],
    "wow" : [],
    "pháº«n ná»™" : []
    
}

for index, row in df.iterrows():
    try:
        row["reactions"] = row["reactions"].replace("'", "\"")
        dict = json.loads(row["reactions"])
        for key, value in reactions_list.items():
            if key in dict:
                reactions_list[key].append(dict[key])
            else:
                reactions_list[key].append(0)
    except:
        for key, value in reactions_list.items():
            reactions_list[key].append(0)
df[["thÃ­ch", "thÆ°Æ¡ng thÆ°Æ¡ng", "yÃªu thÃ­ch", "haha", "buá»“n", "wow", "pháº«n ná»™"]] = pd.DataFrame(reactions_list)
```
