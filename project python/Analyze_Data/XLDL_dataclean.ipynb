xử lí dữ liệu thô

```python
import json
```

```python
# import json
df["reactions"]

reactions_list = {
    "thích" : [],
    "thương thương" : [],
    "yêu thích" : [],
    "haha" : [],
    "buồn" : [],
    "wow" : [],
    "phẫn nộ" : []
    
}

for index, row in df.iterrows():
    try:
        row["reactions"] = row["reactions"].replace("'", "\"")
        dict = json.loads(row["reactions"])
        for key, value in reactions_list.items():
            if key in dict:
                reactions_list[key].append(dict[key])
            else:
                reactions_list[key].append(0)
    except:
        for key, value in reactions_list.items():
            reactions_list[key].append(0)
df[["thích", "thương thương", "yêu thích", "haha", "buồn", "wow", "phẫn nộ"]] = pd.DataFrame(reactions_list)
```
